```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import geopandas as gpd

pd.set_option("max_columns", None)
```

```python
df = pd.read_csv('data/final_data.csv')

df['Cases/100000'] = df['COVID-19 Cases'] / (df['Population']/100000)
df['Malaria Cases/100000'] = df['COVID-19 Cases'] / (df['Population']/100000)
df['Country Type'] = df['Malaria Cases'] > 0
df['Country Type'] = df['Country Type'].replace(True, 'Malarious')
df['Country Type'] = df['Country Type'].replace(False, 'Non-malarious')
df['Flights/100000'] = df['Air Passengers Carried'] / (df['Population']/100000)
df['Tests/100000'] = df['Tests Conducted'] / (df['Population']/100000)
```

```python
corr = df.corr() 
corr['COVID-19 Cases'].abs().sort_values(ascending=False).head(5)
```

```python
corr['Malaria Cases'].abs().sort_values(ascending=False).head(5)
```

```python
average_cases = df[['Country', 'Malaria Cases']].groupby(df['Malaria Cases']>0).count()
average_cases
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
covid_df = df[['Country Code', 'COVID-19 Cases', 'Malaria Cases']]
covid_df = covid_df.fillna(0)

world_covid_df = pd.merge(world, covid_df, left_on='iso_a3', right_on='Country Code')
```

```python
ax = world_covid_df.dropna().plot(column='COVID-19 Cases', cmap ='OrRd', k=5, figsize=(15,9), scheme='quantiles', legend = True)
plt.title('Occurence of COVID-19 Cases in the World')
plt.show()
```

```python
ax = world_covid_df.dropna().plot(column='Malaria Cases', cmap ='OrRd', k=3, figsize=(15,9), scheme='quantiles', legend = True)
plt.title('Occurence of Malaria Cases in the World')
plt.show()
```

```python
x = df.groupby(['Country Type'], as_index=False).mean()
sns.barplot(x='Country Type', y='Cases/100000', data=x, order=['Non-malarious', 'Malarious'], saturation=0.75)
plt.title('Average COVID-19 Cases in Malarious vs Non-malarious Countries')
plt.ylabel('Average COVID-19 Cases/100,000 people')
plt.show()
```

```python
sns.relplot(x="Tests/100000", y="Cases/100000", data=df, hue='Country Type', hue_order=['Non-malarious', 'Malarious']);
plt.title('COVID-19 Cases/100,000 vs Tests Conducted/100,000')
plt.show()
```

```python
s = df[df['Air Passengers Carried'] < 600000000]
sns.relplot(x="Air Passengers Carried", y="COVID-19 Cases", data=s, hue='Country Type', hue_order=['Non-malarious', 'Malarious']);
plt.title('COVID-19 Cases vs Air Passengers Carried')
plt.show()
```

```python
x = df.groupby(['Country Type'], as_index=False).mean()
sns.barplot(x='Country Type', y='Air Passengers Carried', data=x, order=['Non-malarious', 'Malarious'], saturation=0.75)
```

```python
s = df[df['Air Passengers Carried'] < 600000000]
sns.relplot(x="Migrants", y="COVID-19 Cases", data=s, hue="Country Type", hue_order=['Non-malarious', 'Malarious']);
plt.title('COVID-19 Cases vs Number of Migrants')
plt.show()
```

```python
x = df.groupby(['Country Type'], as_index=False).mean()
plt.hlines(0, -3000, 3000)
sns.barplot(x='Country Type', y='Migrants', data=x, order=['Non-malarious', 'Malarious'], saturation=0.75)
plt.title('Average Number of Migrants in Malarious vs Non-malarious Countries')
plt.show()
```
